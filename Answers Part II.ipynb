```python
def anagrams_of(s): 
    # Factorial
        def factorial(ls): 
            if ls == 1: 
                return 1  
            else: 
                return ( ls * factorial(ls-1) )    
        f_s = factorial(len(s)) 
    # Recursion Of One Combination
        len_ls = int( f_s / len(s) )
        def ao(s,ls = len_ls): 
            if ls == 1: 
                return s[1:] + s[:1] 
            else:  
                for j in range(0,len_ls): 
                #print(s[1:] + s[:1] + ao((s[1:] + s[:1]))   ) 
                    return s[j:] + s[:j] + ao((s[1:] + s[:1]), ls-1)    
    # Total Combinations         
        def combinations(x = s, m = len_ls ):
            xl =[x, x[::-1]]        
            x += x   
            for i,j in zip(range(len(x)), range(1,len(x))): 
                t = x[j] + x[i] + x[j+1:j+3] 
                xl.append(t) 
            return xl[:m] 
        comb = combinations()  
        res = ''
        for i in comb:
            res += ao(i) 
        
        n = 4
        res = [res[i:i+n] for i in range(0, len(res), n)] 
        t = []
        for i in res:
            if i not in t:
                t.append(i) 
        return t 
    
print(anagrams_of("pots"))
```
