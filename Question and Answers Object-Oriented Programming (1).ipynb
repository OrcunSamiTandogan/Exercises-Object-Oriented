```python
def substring_count(first, second):
    count = 0 
    try: 
        flen = len(first)
        slen = len(second) 
        for i in range(flen):
            temp = first[i:slen] 
            slen += 1
            if temp == second: 
                count += 1 
        return count         
    except: 
        return 'You have not entered proper string value at least for one variable'
            
    
print(substring_count("lifeforlifeforlifeforlife","life"))  # returns 4
substring_count("lifeforlifeforlifeforlife","lifeforlife") # returns 3
```

Answer V 

```python
import math

class Polygon:
    def __init__(self, sides):
        """Initialized with the lengths of the sides"""
        for s in sides:
            if s <= 0:
                raise ValueError('Side lengths must be positive')
        self._sides = sides
        self._name = "Polygon"

    def name(self):
        return self._name

    def area(self):
        """The area of an arbitrary polygon is unknown"""
        return None #placeholder

    def perimeter(self):
        total = 0.0
        for s in self._sides:
            total += s
        return round(total, 4 )


class Triangle(Polygon): 
    def __init__(self, s1, s2, s3 ): 
        sides = [s1,s2,s3] 
        super().__init__(sides) 
        self._side1 = s1 
        self._side2 = s2 
        self._side3 = s3 
        self._name = 'Triangle'

    def __name__(self):
        return self._name 

    def area(self): 
        a = self._side1 
        b = self._side2 
        c = self._side3 
        s = (a + b + c) / 2   
        self._area = (s*(s-a)*(s-b)*(s-c)) ** 0.5          
        return round(self._area , 4 )  

    def __iter__(self):
        return self         

class IsocelesTriangle(Polygon): 
    def __init__(self, s1, s2 ):  
        sides = [s1*2,s2] 
        super().__init__(sides) 
        self._side1 = s1 
        self._side2 = s2 
        self._name = 'IsocelesTriangle'

    def __name__(self):
        return self._name 
    
    def area(self):
        a = self._side1 
        b = self._side2 
        self._area = (b * math.sqrt((4 * a * a) - (b * b)))/4 
        return round(self._area , 4 ) 


class EquilateralTriangle(Polygon): 
    def __init__(self, s1 ):  # IsocelesTriangle(4.1,1),
        sides = [s1*3] 
        super().__init__(sides) 
        self._side1 = s1 
        self._name = 'EquilateralTriangle' 

    def __name__(self):
        return self._name 
    
    def area(self):
        a = self._side1 
        self._area = (math.sqrt(3)/ 4)*(a * a) 
        return round( self._area , 4 ) 



class Rectangle(Polygon): 
    def __init__(self, s1,s2 ):  # IsocelesTriangle(4.1,1),
        sides = [s1*2,s2*2] 
        super().__init__(sides) 
        self._side1 = s1 
        self._side2 = s2 
        self._name = 'Rectangle' 

    def __name__(self):
        return self._name 
    
    def area(self):
        a = self._side1 
        b = self._side2
        self._area = a*b
        return round(self._area, 4)  



class Square(Polygon): 
    def __init__(self, s1 ):  
        sides = [s1*4] 
        super().__init__(sides) 
        self._side1 = s1 
        self._name = 'Square' 

    def __name__(self):
        return self._name 
    
    def area(self):
        a = self._side1 
        self._area = a*a 
        return round(self._area , 4) 
```
